```python
import pandas as pd
import numpy as np
diabetes = pd.read_csv('diabetes.csv')
htn = pd.read_csv('htn.csv')
stroke = pd.read_csv('stroke.csv')
```

```python
# Links used to make changes:
# https://www.cdc.gov/bloodpressure/about.htm
# https://www.niddk.nih.gov/health-information/diabetes/overview/tests-diagnosis#:~:text=Your%20doctor%20can%20diagnose%20diabetes,type%20of%20diabetes%20you%20have.
```

```python
diabetes.head()
```

```python
columns_drop = ['CholCheck', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'MentHlth', 'PhysHlth', 'DiffWalk', ]
columns_drop2 = ['Age', 'Chest_Pain', 'Rest_BP', 'Rest_ECG', 'Max_HeartRate', 'ExAng', 'OldPeak', 'Slope', 'CA', 'Thalassemia']
columns_drop3 = ['Age', 'Ever_Married', 'Residence_type']
```

```python
# diabetes.drop(columns_drop, axis=1, inplace=True)
diabetes = diabetes.sort_index(axis=1)
diabetes.info()
```

```python
diabetes.rename(columns={'HighBP': 'Hypertension'}, inplace=True)
```

```python
diabetes['Fast_Blood_Sugar'] = np.nan
diabetes['Work_Type'] = np.nan
diabetes['Avg_Glucose_Level'] = np.nan
diabetes['Work_Type'] = np.nan
diabetes['Age_Group'] = diabetes['Age_Group'].astype(object)
```

```python
diabetes.head()
```

```python
# htn.drop(columns_drop2, axis=1, inplace=True)
htn = htn.sort_index(axis=1)
htn.info()
```

```python
#done 
htn['Work_Type'] = np.nan
htn['Smoker'] = np.nan
htn['HeartDiseaseorAttack'] = np.nan
htn['PhysActivity'] = np.nan
htn['GenHlth'] = np.nan
htn['Stroke'] = np.nan
htn['Diabetes'] = np.nan
htn['BMI'] = np.nan
htn['Avg_Glucose_Level'] = np.nan
htn['BMI'] = np.nan
htn['HighChol']=htn['HighChol'].astype(float)
htn['Fast_Blood_Sugar']=htn['Fast_Blood_Sugar'].astype(float)
htn['Hypertension']=htn['Hypertension'].astype(float)
htn['Age_Group']=htn['Age_Group'].astype(object)
```

```python
#done
stroke['Smoker'] = stroke['Smoker'].astype(float)
stroke['Heart_Disease'] = stroke['Heart_Disease'].astype(float)
stroke['Stroke'] = stroke['Stroke'].astype(float)
stroke['Work_Type'] = stroke['Work_Type'].astype(float)
stroke['Hypertension'] = stroke['Hypertension'].astype(float)
stroke['Fast_Blood_Sugar'] = np.nan
stroke['HighChol'] = np.nan
stroke['PhysActivity'] = np.nan
stroke['GenHlth'] = np.nan
stroke['Diabetes'] = np.nan
```

```python
htn.rename(columns={'Chol': 'HighChol',}, inplace=True)
htn.drop('Chol', axis=1, inplace=True)
```

```python
htn['HighChol'] = np.where(htn['Chol'] > 200, 1, 0)
```

```python
# stroke.drop(columns_drop3, axis=1, inplace=True)
stroke = stroke.sort_index(axis=1)
stroke.info()
```

```python
stroke.rename(columns={'Heart_Disease': 'HeartDiseaseorAttack'}, inplace=True)
```

```python
concatenated = pd.concat([diabetes, htn, stroke], axis=0, ignore_index=True)
```
